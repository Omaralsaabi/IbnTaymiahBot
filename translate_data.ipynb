```python
from transformers import MarianMTModel, MarianTokenizer
import pandas as pd
```

```python
df = pd.read_csv('data/مجموع الفتاوى_cleaned_cropped.csv', encoding='utf-8')

df.head()
```

```python
cleaned_questions = df['cleaned'].tolist()

model_name = 'Helsinki-NLP/opus-mt-ar-en'
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

batch_size = 8  # You can adjust this based on your available resources
translated_questions = []

for i in range(0, len(cleaned_questions), batch_size):
    batch = cleaned_questions[i:i+batch_size]
    inputs = tokenizer(batch, return_tensors="pt", padding=True, truncation=True, max_length=512)
    outputs = model.generate(**inputs)
    decoded_outputs = [tokenizer.decode(output, skip_special_tokens=True) for output in outputs]
    translated_questions.extend(decoded_outputs)

df['translated_question'] = translated_questions
```
